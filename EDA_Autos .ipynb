<h3 style= "color: #2F4F4F">2.1 Identificacion de Valores faltantes o perdidos y estandarizacion</h3>

```python
dfsetAutos
```

```python
dfsetAutos.replace("?", np.nan , inplace = True)
```

```python
dfsetAutos
```


<h3 style= "color: #2F4F4F">2.2 Contar los valores NaN en el set de datos</h3>

```python
valoresPerdidos = dfsetAutos.isnull()
valoresPerdidos.head(5)
```

```python
for column in valoresPerdidos.columns.values.tolist():
    print (column)
    print(valoresPerdidos[column].value_counts())
    print ("")
```

<h3 style= "color: #2F4F4F">2.3 Imputar valores numericos con la media</h3>

```python
avg_norm = dfsetAutos["normalized"].astype("float").mean(axis=0)
print("media del normalized:", avg_norm)
```

```python
dfsetAutos["normalized"].replace(np.nan, 124 , inplace = True )
dfsetAutos.head(5)
```

<h3 style= "color: #2F4F4F">2.4 Imputar valores STRING</h3>

```python
dfsetAutos["doors"].value_counts().idxmax()
```

```python
dfsetAutos["normalized"].replace(np.nan, "four" , inplace = True )
dfsetAutos.head(5)
```

<h3 style= "color: #2F4F4F">2.5 Eiminar filas con valores o registros perdidos</h3>
